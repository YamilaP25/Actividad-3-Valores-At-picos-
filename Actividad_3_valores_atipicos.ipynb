#1. Método aplicando cuartiles para la columna ventas_precios_corrientes

```python
#Realkizamos diagrama de caja o bigote
fig = plt.figure(figsize = (5, 3))
plt.boxplot(df1['ventas_precios_corrientes'])
plt.title('Outliers de ventas precios corrientes')
plt.show() #Dibujamos el histograma
```

```python
y = df1['ventas_precios_corrientes']

percentile25 = y.quantile(0.25) #Q1
percentile75 = y.quantile(0.75) #Q2
iqr = percentile75 - percentile25

Limite_Superior_iqr = percentile75 + 1.5*iqr
Limite_Inferior_iqr = percentile25 - 1.5*iqr
print('Límite superior permitido:', Limite_Superior_iqr)
print('Límite inferior permitido:', Limite_Inferior_iqr)
```

```python
outliers_iqr = df1[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers_iqr
```

```python
df_clean_ventas_precios_corrientes = df1[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
df_clean_ventas_precios_corrientes #los elimina
```

```python
df_clean_ventas_precios_corrientes.to_csv('clean_ventas_precios_corrientes.csv')
```

```python
#Realizamos diagrama de caja o bigote
fig = plt.figure(figsize = (5, 3))
plt.boxplot(df_clean_ventas_precios_corrientes['ventas_precios_corrientes'])
plt.title('Outliers de ventas precios corrientes')
plt.show() #Dibujamos el histograma
```

#2. Método de desviación estándar para la columna ventas_precios_constantes

```python
#Realizamos diagrama de caja o bigote
fig = plt.figure(figsize = (5, 3))
plt.boxplot(df1['ventas_precios_constantes'])
plt.title('Outliers de ventas precios constantes')
plt.show() #Dibujamos el histograma
```

```python
#Método aplicando desviación estándar. Encuentro los valores extremos
y = df1['ventas_precios_constantes']
Limite_Superior = y.mean() + 3*y.std()
Limite_Inferior = y.mean() - 3*y.std()
print('Límite superior permitido:', Limite_Superior)
print('Límite inferior permitido:', Limite_Inferior)
```

```python
#Encontrammos Outliers
outliers = df1[(y>Limite_Superior)|(y<Limite_Inferior)]
outliers
```

```python
#obtenemos datos limpios
df_clean_ventas_precios_constantes = df1[(y<=Limite_Superior)&(y>=Limite_Inferior)]
df_clean_ventas_precios_constantes #los elimina
```

```python
df_clean_ventas_precios_constantes.to_csv('clean_ventas_precios_constantes.csv')
```
