#1. Hoja 2020

```python
df = pd.read_excel('Gastos y costos 20-23.xlsx', sheet_name = '2020', skiprows = 5)
df.head(5)
```

```python
df.info()
```

```python
valores_nulos = df.isnull().sum()
valores_nulos
```

```python
#Reemplazamos los valores nulos del dataFrame con 'bfill'
df1 = df.fillna(method = 'bfill')
df1
```

```python
#Reemplazamos los valores nulos del dataFrame con 'ffill'
df1 = df1.fillna(method = 'ffill')
df1
```

```python
valores_nulos = df1.isnull().sum()
valores_nulos
```

```python
df.info()
```

```python
cuantitativas = df1.iloc[ : , [8, 9, 10, 11, 12, 13, 14, 18]]
cualitativas = df1.iloc[ : , [0, 1, 2, 3, 4, 5, 6, 7, 15, 16, 17]]
cuantitativas
```

```python
#Realizamos diagrama de caja o bigote de cada columna del dataFrame
fig = plt.figure(figsize = (15, 8))
cuantitativas.plot(kind = 'box', vert = False)
plt.title('Valores Atípicos del DataFrame')
plt.show() #Se dibuja el diagrama
```

```python
#Método de desviación estándar para eliminar outliers en DataFrame
y = cuantitativas
Limite_Superior = y.mean() + 3*y.std()
Limite_Inferior = y.mean() - 3*y.std()
print('Límite superior permitido:', Limite_Superior)
print('Límite inferior permitido:', Limite_Inferior)
```

```python
#Encontrammos Outliers
outliers = cuantitativas[(y>Limite_Superior)|(y<Limite_Inferior)]
outliers
```

```python
#obtenemos datos y los outliers se conveoerten en nulos en el dataFrame
df2 = cuantitativas[(y <= Limite_Superior) & (y >= Limite_Inferior)]
df2
```

```python
valores_nulos = df2.isnull().sum()
valores_nulos
```

```python
#Reemplaza valores atípicos (nulos) del dataframe con el promedio 'mean'
#Realizamos una copia del dataFrame
df_clean = df2.copy()
df_clean = df_clean.fillna(round(df2.mean(),1))
df_clean
```

```python
valores_nulos = df_clean.isnull().sum()
valores_nulos
```

```python
Datos_limpios = pd.concat([cualitativas, df_clean], axis = 1)
Datos_limpios
```

```python
Datos_limpios.to_csv('Datos_limpios_2020.csv')
```

#3. Hoja 2022

```python
df = pd.read_excel('Gastos y costos 20-23.xlsx', sheet_name = '2022', skiprows = 3)
df = df.drop(['Otros'], axis = 1)

df.head(5)
```

```python
df.info()
```

```python
#Reemplazamos los valores nulos del dataFrame con 'bfill'
valores_nulos = df.isnull().sum()
print(valores_nulos)
df1 = df.fillna(method = 'bfill')
df1 = df1.fillna(method = 'ffill')
valores_nulos = df1.isnull().sum()
print('\n', valores_nulos)
```

```python
#Realizamos diagrama de caja o bigote de cada columna del dataFrame
fig = plt.figure(figsize = (15, 8))
cuantitativas.plot(kind = 'box', vert = False)
plt.title('Valores Atípicos del DataFrame')
plt.show() #Se dibuja el diagrama
```

```python
cuantitativas = df1.iloc[ : , [8, 9, 10, 11, 12, 13, 14]]
cualitativas = df1.iloc[ : , [0, 1, 2, 3, 4, 5, 6, 7, 15, 16, 17]]
cuantitativas
```

```python
y = cuantitativas

percentile25 = y.quantile(0.25) #Q1
percentile75 = y.quantile(0.75) #Q2
iqr = percentile75 - percentile25

Limite_Superior_iqr = percentile75 + 1.5 * iqr
Limite_Inferior_iqr = percentile25 - 1.5 * iqr
print('Límite superior permitido:', Limite_Superior_iqr)
print('Límite inferior permitido:', Limite_Inferior_iqr)
```

```python
outliers_iqr = cuantitativas[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers_iqr
```

```python
#obtenemos datos limpios
df3_iqr = cuantitativas[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
df3_iqr
```

```python
valores_nulos = df3_iqr.isnull().sum()
valores_nulos
```

```python
df_clean_iqr =df3_iqr.copy()
df_clean_iqr =df3_iqr.fillna(round(df3_iqr.mean(),1))
df_clean_iqr
```

```python
valores_nulos = df_clean_iqr.isnull().sum()
valores_nulos
```

```python
Datos_limpios = pd.concat([cualitativas, df_clean_iqr], axis = 1)
Datos_limpios
```

```python
Datos_limpios.to_csv('Datos_limpios_2022.csv')
```
